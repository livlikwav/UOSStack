```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# 영어 문자들만 남기는거 > 정규 표현식 활용함
```

```python
letters_only = re.sub('[^a-zA-Z]', ' ', line)
# re.sub 정규표현식으로 substitute(대체) 한다. line안에 있는거, ' 공백 ' 으로
```

```python
letters_only[:20]
```

```python
letters_only = letters_only.lower()
```

```python
token = nl.word_tokenize(letters_only)
```

```python
len(token)
```

```python
vocab = nl.Text(token).vocab()
```

```python
vocab
```

```python
len(token)
```

```python
# a나 s 같은게 생기는데 이런거 없애버리자 일단은 1글자 짜리!

for i in range(0, len(token)-1):
    if(len(token[i])==1):
        del token[i]
```

```python
words = [w for w in token if not w in stopwords.words('english')]
```

## wordcloud 생성자에 파라미터 넣어서 생성함
가장 간단한 word cloud 를 그리는 코드는 아래와 같습니다. generate_from_text() 는 텍스트에서 단어의 빈도를 계산하여 이를 기반으로 클라우드를 그리며, generate_from_frequencies() 는 미리 정의된 단어의 빈도수 혹은 중요도를 이용하여 클라우드를 그립니다.

<https://lovit.github.io/nlp/2018/04/17/word_cloud/>

## plt.imshow() 이미지 그리는 pyplot
Imshow
화상(image) 데이터처럼 행과 열을 가진 행렬 형태의 2차원 데이터는 imshow 명령을 써서 2차원 자료의 크기를 색깔로 표시하는 것이다.

데이터 수치를 색으로 바꾸는 함수는 칼라맵(color map)이라고 한다. 칼라맵은 cmap 인수로 지정한다. 사용할 수 있는 칼라맵은 plt.cm의 속성으로 포함되어 있다. 아래에 일부 칼라맵을 표시하였다. 칼라맵은 문자열로 지정해도 된다. 칼라맵에 대한 자세한 내용은 다음 웹사이트를 참조한다.

```python
word_cloud = WordCloud(width=2000, height=1000, background_color="white", random_state=0)
word_cloud = word_cloud.generate_from_frequencies(vocabs)
word_cloud
```

```python
plt.figure(figsize=(12,12))
plt.imshow(word_cloud)
plt.axis("off")
plt.show 
```
